```python
import random

class Chatbot:
    def __init__(self):
        self.previous_context = {}

    def greet(self):
        responses = [
            "Hello! I'm your friendly chatbot. How can I assist you today?",
            "Hi there! Welcome to the chatbot. What can I do for you?",
            "Hey! Nice to see you here. How may I help you?"
        ]
        print(random.choice(responses))

    def respond_to_basic_questions(self, question):
        responses = {
            "What is your name?": "My name is Chatbot!",
            "How are you?": "I'm just a computer program, but I'm functioning well! How about you?",
            "What can you do?": "I can assist you with various tasks and provide information. Just ask!",
            "Where are you from?": "I exist in the digital world, so you can find me everywhere!",
            "Do you like programming?": "As a chatbot, I don't have personal preferences, but I enjoy assisting users!"
        }
        if question in responses:
            return responses[question]
        else:
            return "I'm sorry, I don't have an answer to that question."

    def ask_questions(self):
        questions = [
            "What's your favorite color?",
            "How was your day today?",
            "What do you like to do in your free time?"
        ]
        for question in questions:
            answer = input(f"Chatbot: {question}\nYou: ").strip().lower()
            self.previous_context[question] = answer
            print("Chatbot: That sounds interesting!")

    def farewell(self):
        print("Chatbot: It was nice chatting with you! Goodbye!")

    def handle_input(self, user_input):
        if "exit" in user_input:
            self.farewell()
            return False
        elif "?" in user_input:
            response = self.respond_to_basic_questions(user_input)
            print(f"Chatbot: {response}")
        else:
            print("Chatbot: I'm sorry, I didn't understand that. Can you please rephrase or ask something else?")
        return True

    def chat(self):
        self.greet()
        continue_chat = True
        while continue_chat:
            user_input = input("You: ").strip()
            continue_chat = self.handle_input(user_input)

if __name__ == "__main__":
    chatbot = Chatbot()
    chatbot.chat()
```
